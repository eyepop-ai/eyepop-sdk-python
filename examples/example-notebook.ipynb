```python
import os
from eyepop import EyePopSdk

from dotenv import load_dotenv

# Use dotenv to automatically load required env vars...
load_dotenv()

# ...or set yourself!
eyepop_url = os.getenv('EYEPOP_URL')

def upload_photo(file_path: str):
    with EyePopSdk.workerEndpoint(eyepop_url=eyepop_url) as endpoint:
        result = endpoint.upload(file_path).predict()
        print(result)
        assert result.get('objects') is not None


upload_photo('examples/example.jpg')
```
